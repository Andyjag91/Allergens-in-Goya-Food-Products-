```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
from time import sleep
```

```python
url='https://world.openfoodfacts.org/brand/goya/ingredients'
response=requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#code to do the following
#1.get all the ingredient list
#2.get all the links to access the products


link_list=[]
ingredient_list=[]
results = soup.find_all("a", {'class':["tag known","tag user_defined"]})
url='https://world.openfoodfacts.org'
for result in results:
    try:   
        sleep(0.5)
        ingredients=result.text
        link= result["href"]
        print(link)
        base_url=url+link
        print(base_url)
        link_list.append(base_url)
        ingredient_list.append(ingredients)
        
    except:
        pass
```

```python
#getting all the values of ingredients having milk,peanut,soybean,wheat,tree-nut,egg,shellfish and fish
subset_list = [text for text in link_list if any(k in text for k in ['milk', 'peanut', 'soybean','wheat','tree-nut','egg','shellfish','fish'])]
#list2 = [text for text in list1 if any(k in text for k in ['a', 'b', 'c'])]
subset_list
```

```python
#using all the ingre urls getting all the products having ingre and storing in subset_dict
subset_dict={"Ingredient":[],"Products":[]}
pro_list=[]
for link in subset_list:
    try:
        ingredient = link.split('/')[-1]
        response=requests.get(link)
        value_soup= BeautifulSoup(response.text, 'html.parser')
        values=value_soup.find('div', id="search_results").find_all('a')
        for value in values:
        #print(value)
            pro_list.append(value["title"])
            print(pro_list)
    
        subset_dict["Ingredient"].append(ingredient)
        subset_dict["Products"].append(pro_list)
        pro_list=[]
    except:
        pass
```

```python
final_subset_df=pd.DataFrame.from_dict(subset_dict)
```

```python
final_subset_df.to_csv("final_subset.csv")
```
