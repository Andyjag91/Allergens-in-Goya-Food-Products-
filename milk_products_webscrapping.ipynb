```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
from time import sleep
```

```python
url='https://world.openfoodfacts.org/brand/goya/ingredients'
response=requests.get(url)
```

```python
#print(response.text)
soup = BeautifulSoup(response.text, 'html.parser')
results = soup.find_all("a", class_="tag known")
```

```python
#code to do the following
#1.get all the ingredient list
#2.get all the links to access the products
#3.get all the products and respective ingredients in a dict

final_dict={"Ingredient":[],"Products":[]}
product_list=[]
link_list=[]
ingredient_list=[]
results = soup.find_all("a", class_="tag known")
url='https://world.openfoodfacts.org'


for result in results:
    try:   
        sleep(0.5)
        ingredients=result.text
        link= result["href"]
        base_url=url+link
        link_list.append(base_url)
        
        response=requests.get(base_url)
        value_soup= BeautifulSoup(response.text, 'html.parser')
        values=value_soup.find('div', id="search_results").find_all('a')

        for value in values:
            #print(value)
            product_list.append(value["title"])
            #print(product_list)
             
        ingredient_list.append(ingredients)
        final_dict["Ingredient"].append(ingredients)
        final_dict["Products"].append(product_list)
        product_list=[]
    except:
        pass
```

```python
#getting all the values of ingredients having milk
milk_ingre = [x for x in ingredient_list if "milk" in x]
milk_ingre
```

```python
#replacing all the spaces with "-" in the ingredient list
new_milk_list=[]
for i in milk_ingre:
    j = i.replace(' ','-')
    new_milk_list.append(j)
  
print(new_milk_list)
```

```python
#Finally getting all the ingredients and products having milk in a dictionary  

milk_dict={"Ingredient":[],"Products":[]}
pro_list=[]
url="https://world.openfoodfacts.org/brand/goya/ingredient/"
for ingre in new_milk_list:
    print(ingre)
    base_url=url+ingre
    print(base_url)
    
    response=requests.get(base_url)
    value_soup= BeautifulSoup(response.text, 'html.parser')
    values=value_soup.find('div', id="search_results").find_all('a')
    for value in values:
        print(value)
        pro_list.append(value["title"])
        print(product_list)
    
    milk_dict["Ingredient"].append(ingre)
    milk_dict["Products"].append(pro_list)
    pro_list=[]
```

```python
#converting to dataframe
final_milk_df=pd.DataFrame.from_dict(milk_dict)
```
