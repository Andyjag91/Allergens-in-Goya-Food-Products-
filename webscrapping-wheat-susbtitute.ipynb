```python
url="https://www.healthline.com/nutrition/10-ways-to-replace-bread#TOC_TITLE_HDR_2"
response=requests.get(url)
```

```python
soup=BeautifulSoup(response.text,'html.parser')
print(soup.prettify())
```

```python
results=soup.find_all( class_="chartbeat-section")
print(results)
```
